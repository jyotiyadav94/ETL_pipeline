## Autoviz

```python
import pandas as pd
from autoviz import AutoViz_Class
%matplotlib inline

AV = AutoViz_Class()

data = pd.read_csv('/content/drive/MyDrive/company cherrry/transformed.csv')
df = pd.DataFrame(data)

dft = AV.AutoViz(
    "",
    sep=",",
    depVar="",
    dfte=df,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir="/content/Autoviz"
)
```

```python
# Filter rows where 'particella' and 'subalterno' contain characters other than digits and letters
# Convert 'particella' and 'subalterno' to string and then to integer (if possible)
df['particella'] = df['particella'].astype(str).str.split('.').str[0].astype(str)
df['subalterno'] = df['subalterno'].astype(str).str.split('.').str[0].astype(str)

# Check if 'particella' and 'subalterno' contain characters other than digits and letters
filtered_df = df[~df['particella'].str.match('^[a-zA-Z0-9]*$') | ~df['subalterno'].str.match('^[a-zA-Z0-9]*$')]
filtered_df
```

```python
# checking duplicates

duplicate_rows = df[df.duplicated()]

if duplicate_rows.empty:
    print("No duplicates found in the DataFrame.")
else:
    print("Duplicate rows found in the DataFrame:")
    print(duplicate_rows)
```

```python
duplicates = df[df.duplicated(subset=['cherry_asset_id'], keep=False)]

if duplicates.empty:
    print("No duplicates found in the column.")
else:
    print("Duplicate values found in the column:")
duplicates
```

```python
def delete_nan_rows(df):
    """
    Delete rows where both 'vatCode' and 'taxCode' are NaN.
    """
    cleaned_df = df.dropna(subset=['vatCode', 'taxCode'], how='all')
    return cleaned_df


def remove_duplicates(df):
    """
    Remove duplicate entries.
    """
    return df.drop_duplicates()
```
